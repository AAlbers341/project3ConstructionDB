```python
import pandas as pd
```

```python
sizes_df = pd.read_csv('../../data/raw/sizes.csv')
sizes_df
```

```python
sizes_df.columns
```

```python
# Dropping 'metric' columns. The original estimate softare was in meters 'metric'
# repopulated with feet 'Imperial'
sizes_df = sizes_df.drop(['DimDepthMetric',
                          'DimWidthMetric',
                          'DimThicknessMetric',
                          'DimGageMetric',
                          'DimDiameterMetric',
                          'DimensionSizesMetric',
                          'DimensionSizesMetric_Reverse',
                         'DimensionSizesMetric_EDI',
                         'Dim1Metric',
                         'Dim2Metric',
                         'Dim3Metric',
                         'Dim4Metric',
                         'Dim5Metric',
                         'Dim6Metric'], axis=1)

# See if columns were dropped successfully
sizes_df.columns
```

```python
sizes_df.to_csv('../../data/cleaned/sizes.csv', index=False)
```
